```python
import openpyxl
```

```python
from bs4 import BeautifulSoup
```

```python
import requests

```

```python
url = "https://elitka.kg/object/index?query=&city%5B%5D=2&finish_year=&object_class%5B%5D=2"


```

```python
response = requests.get(url)
html = response.text
```

```python
bs = BeautifulSoup(response.text,"lxml")
print(bs)
```

```python
addresses = bs.find_all('div', class_='obj-address')
```

```python
prices = bs.find_all('div', class_='obj-price')

```

```python

if len(addresses) == len(prices):
    data = []

    for i in range(len(addresses)):
        address = addresses[i].text
        price_value = prices[i].text


        data.append([address, price_value])


    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Данные"


    sheet.append(['Адрес', 'Цена'])
    for row in data:
        sheet.append(row)


    workbook.save('output.xlsx')

    print("Данные успешно сохранены в output.xlsx.")
else:
    print("Количество элементов с классами obj-address и obj-price не совпадает.")
```
